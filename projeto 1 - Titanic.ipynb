```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df_result = df[['Name','Age', 'Sex', 'Pclass', 'Survived']].loc[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)].sort_values(['Name'])
df_result
```

```python
df_result.to_csv('titanic_results.csv', index = False)
```
